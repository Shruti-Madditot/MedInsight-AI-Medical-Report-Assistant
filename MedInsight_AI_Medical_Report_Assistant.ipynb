```python
text = text.replace("\n", " ")
text = text.strip()

print("Text Cleaned Successfully")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100
)

chunks = text_splitter.split_text(text)

print("Total Chunks:", len(chunks))
```

```python
chunks[:2]
```

#  Create Embeddings

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("✅ Embedding Model Loaded Successfully")
```

# Create FAISS Vector Database

```python
vector_db = FAISS.from_texts(
    texts=chunks,
    embedding=embedding_model
)

print("FAISS Vector Database Created Successfully")
```

# Semantic Search Test

```python
query = "What is the patient's cholesterol level?"

results = vector_db.similarity_search(query, k=2)

print("Retrieved Documents:\n")

for i, doc in enumerate(results, start=1):
    print(f"Result {i}")
    print(doc.page_content)
    print("-" * 70)
```

#  AI Medical Report Assistant

```python
retriever = vector_db.as_retriever(
    search_kwargs={"k": 2}
)

print(" Retriever Created Successfully")
```

```python
docs = retriever.invoke("What is the patient's blood sugar?")

for doc in docs:
    print(doc.page_content)
```

# AI Medical Question Answering

```python
def medical_chat(question):

    docs = retriever.invoke(question)

    context = "\n\n".join([doc.page_content for doc in docs])

    prompt = f"""
You are an intelligent medical assistant.

Use ONLY the medical report below to answer.

Medical Report:
{context}

Question:
{question}

Answer in simple English.
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
answer = medical_chat(
    "What is the patient's cholesterol level?"
)

print(answer)
```

```python
answer = medical_chat(
    "What is the patient's Vitamin D level?"
)

print(answer)
```

```python
answer = medical_chat(
    "Is the patient's blood sugar normal?"
)

print(answer)
```

# Agentic AI Tools

```python
@tool
def summarize_report():
    """
    Generate a summary of the medical report.
    """

    context = "\n".join(chunks)

    prompt = f"""
Summarize the following medical report in simple English.

Medical Report:
{context}
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
print(summarize_report.invoke({}))
```

```python
@tool
def explain_term(term: str):
    """
    Explain a medical term in simple language.
    """

    prompt = f"""
Explain the medical term:

{term}

Use simple English.
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
print(explain_term.invoke("LDL Cholesterol"))
```

```python
@tool
def health_tips():
    """
    Generate health tips based on the report.
    """

    context = "\n".join(chunks)

    prompt = f"""
Based on this report, provide 5 practical health tips.

Medical Report:
{context}
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
print(health_tips.invoke({}))
```

```python
@tool
def doctor_questions():
    """
    Generate questions the patient can ask the doctor.
    """

    context = "\n".join(chunks)

    prompt = f"""
Based on this medical report, suggest 5 questions the patient should ask the doctor.

Medical Report:
{context}
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
print(doctor_questions.invoke({}))
```

```python
@tool
def abnormal_values():
    """
    Detect abnormal values in the report.
    """

    context = "\n".join(chunks)

    prompt = f"""
Find all abnormal or borderline values in this medical report.

Explain why they are abnormal.

Medical Report:
{context}
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
print(abnormal_values.invoke({}))
```

# Agentic AI Medical Assistant

```python
tools = [
    summarize_report,
    explain_term,
    health_tips,
    doctor_questions,
    abnormal_values
]

print("Total Tools:", len(tools))
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.agents import AgentExecutor, create_tool_calling_agent

print(" Agent Modules Imported")
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """
You are MedInsight AI.

You are an intelligent medical assistant.

Always use the available tools whenever needed.

Answer in simple English.
"""
        ),
        ("human", "{input}"),
        ("placeholder", "{agent_scratchpad}")
    ]
)

print(" Prompt Created")
```

```python
agent = create_tool_calling_agent(
    llm=llm,
    tools=tools,
    prompt=prompt
)

print(" Tool Calling Agent Created")
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True
)

print(" Agent Executor Ready")
```

```python
response = agent_executor.invoke(
    {
        "input":"Summarize my medical report"
    }
)

print(response["output"])
```

```python
response = agent_executor.invoke(
    {
        "input":"Explain Vitamin D deficiency"
    }
)

print(response["output"])
```

# MedInsight AI Agent

```python
def medical_agent(user_query):

    query = user_query.lower()

    # Report Summary
    if "summary" in query or "summarize" in query:
        return summarize_report.invoke({"input": user_query})

    # Explain Medical Terms
    elif "explain" in query:
        term = user_query.replace("Explain", "").replace("explain", "").strip()

        if term == "":
            return "Please specify the medical term."

        return explain_term.invoke(term)

    # Health Tips
    elif "health" in query or "tips" in query:
        return health_tips.invoke({"input": user_query})

    # Doctor Questions
    elif "doctor" in query or "question" in query:
        return doctor_questions.invoke({"input": user_query})

    # Abnormal Values
    elif "abnormal" in query or "borderline" in query:
        return abnormal_values.invoke({"input": user_query})

    # Default → Medical Chat (RAG)
    else:
        return medical_chat(user_query)
```
